```python
## # Q.2.Are children more likely to be attacked by sharks?
# Function to categorize age into groups
def categorize_age(age):
    if age < 12:
        return '<12'
    elif 12 <= age < 18:
        return '12-18'
    else:
        return 'Adult'

# Convert 'Age' to numeric  and drop rows with non-numeric values
df3['Age'] = pd.to_numeric(df3['Age'], errors='coerce')
# Impute missing values in 'Age' column with the median age
median_age = df3['Age'].median()
df3['Age'].fillna(median_age, inplace=True)
# Categorize ages into groups using the 'categorize_age' function
df3['Age Group'] = df3['Age'].apply(categorize_age)

# Count the occurrences of each age group
age_group_counts = df3['Age Group'].value_counts()
display(age_group_counts)

# Plotting the data on a bar chart
plt.bar(age_group_counts.index, age_group_counts.values)
plt.xlabel('Age group')
plt.ylabel('Number of Attacks')
plt.title('Shark Attacks by Age Group')

plt.show()
```

* Q.2 .Are children more likely to be attacked by sharks?*
**Answer**: Based on the provided dataset, i analyzed the number of shark attacks involving children (age < 18) and adults (age >= 18). The analysis showed the number of attacks for each age group.
**Assumptions**: The dataset includes information on the age of the victims, but it does not provide additional contextual factors, such as the popularity of certain activities among children, which may influence the likelihood of shark attacks on this age group.

```python
# # 3. Are shark attacks where sharks were provoked more or less dangerous?# Grouping by provoked attacks and counting fatal attacks
#  count all occurrences of provoked and unprovoked in the 'Type' column
Provoked_count = df3[df3['Type'].str.contains('Provoked', na=False)]['Type'].count()
Unprovoked_count = df3[df3['Type'].str.contains('Unprovoked', na=False)]['Type'].count()
print("Number of provoked attacks:", Provoked_count)
print("Number of Unprovoked attacks:", Unprovoked_count)
#count other type also for make
Questionable_count = df3[df3['Type'].str.contains('Questionable', na=False)]['Type'].count()
Invalid_count = df3[df3['Type'].str.contains('Invalid', na=False)]['Type'].count()
print("Number of Invlaid attacks:", Invalid_count)
print("Number of Questionable attacks:", Questionable_count)

# Plotting the data on a bar chart
types = ['Provoked', 'Unprovoked', 'Questionable', 'Invalid']
counts = [Provoked_count, Unprovoked_count, Questionable_count, Invalid_count]

plt.bar(types, counts)
plt.xlabel('Type of Shark Attack')
plt.ylabel('Number of Occurrences')
plt.title('Occurrences of Different Types of Shark Attacks')
plt.show()
```

```python
# # Grouping by provoked  and unprovoked attacks and counting fatal attacks
# Function to categorize attack types as provoked or unprovoked
def categorize_attack_type(attack_type):
    if 'Provoked' in'Attack Type':
        return 'Provoked'
    else:
        return 'Unprovoked'

# Apply the 'categorize_attack_type' function to create a new column 'Attack Type'
df3['Attack Type'] = df3['Type'].apply(categorize_attack_type)

# Count the number of fatal and non-fatal attacks for each attack type
attack_counts = df3.groupby(['Attack Type', 'Fatal (Y/N)'])['Case Number'].count().unstack().fillna(0)
display(attack_counts)
# Plotting the data on a stacked bar chart
plt.figure(figsize=(10, 6))
attack_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.title('Shark Attacks: Provoked vs. Unprovoked')
plt.legend(title='Fatal (Y/N)', loc='upper right', labels=['Non-Fatal', 'Fatal'])
plt.xticks(rotation=0)

plt.show()
```

**Answer**: To answer this question, analyzing the outcomes (fatal or non-fatal) of provoked and unprovoked shark attacks separately.
```python
## 4. Are certain activities more likely to result in a shark attack?
# Impute missing values in 'Activity' column with the most frequent activity (mode)
most_frequent_activity = df3['Activity'].mode().iloc[0]
df3['Activity'].fillna(most_frequent_activity, inplace=True)
# Count the occurrences of each activity
activity_counts = df3['Activity'].value_counts()
## (visualisation is not possible for each activity due to too many sort of activities ).

# Select the top 10 activities with the highest number of shark attacks
top_10 = 10
top_activities = activity_counts.head(top_10)

# Plotting the data on a bar chart
plt.figure(figsize=(12, 6))
top_activities.plot(kind='bar', color='skyblue')
plt.xlabel('Activity')
plt.ylabel('Number of Attacks')
plt.title('Top {} Activities Most Likely to Result in Shark Attacks'.format(top_10))
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()






```

```python
# Count the occurrences of each activity in the particular most happened location
most_happened_location = df3['Location'].value_counts().idxmax()
activities_in_most_happened_location = df3[df3['Location'] == most_happened_location]['Activity'].value_counts()

# Select the top 10 activities with the highest number of shark attacks in the most happened location
top_10 = 10
top_activities = activities_in_most_happened_location.head(top_10)

# Plotting the data on a bar chart
plt.figure(figsize=(12, 6))
top_activities.plot(kind='bar', color='skyblue')
plt.xlabel('Activity')
plt.ylabel('Number of Attacks')
plt.title('Top {} Activities Most Likely to Result in Shark Attacks in {}'.format(top_10, most_happened_location))
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
